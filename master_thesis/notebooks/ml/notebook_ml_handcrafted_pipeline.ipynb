```python
# Training dataset preparation
train_df = pd.read_csv('../../data/training/train_set.csv')
test_df = pd.read_csv('../../data/training/test_set.csv')

df_handcrafted = pd.read_csv('../../data/features/extracted_handcrafted_features.csv', index_col=0)
df_handcrafted['catalogNumber'] = df_handcrafted.index.str.split("/").str[-1].str.split(".").str[0]
df_label_features = pd.read_csv('../../data/features/label_features_normalized.csv', index_col=0)
df_label_features['catalogNumber'] = df_label_features.index.str.split("/").str[-1].str.split(".").str[0]
df_features = df_handcrafted.merge(df_label_features, on='catalogNumber', how='inner')
label_encoder = LabelEncoder()
labels = pd.read_csv('../../data/training/training_dataset_labels.csv')
labels = labels.dropna(subset=['best_tool'])
label_encoder.fit(labels['best_tool'])
label_mapping = {index: label for index, label in enumerate(label_encoder.classes_)}
training_dataset_without_manual_review = pd.read_csv('../../data/training/training_dataset_labels_wo_manual_review.csv')
# training_dataset_without_manual_review = pd.read_csv('../../data/training/training_dataset_labels.csv')

df_features_labels = df_features.merge(labels[['catalogNumber', 'best_tool']], on='catalogNumber', how='inner')
df_features_labels['best_tool_encoded'] = label_encoder.transform(df_features_labels['best_tool'])
df_features_labels = df_features_labels.drop(columns=['best_tool'])
df_features_labels = df_features_labels.set_index('catalogNumber')
x_train_handcrafted = df_features_labels.loc[train_df['catalogNumber']].drop(columns=['best_tool_encoded'])
y_train_handcrafted = df_features_labels.loc[train_df['catalogNumber'], 'best_tool_encoded']
# store trainig set
df_features_labels.loc[train_df['catalogNumber']].to_csv('../../data/training/train_set_features.csv')
# store test set
df_features_labels.loc[test_df['catalogNumber']].to_csv('../../data/training/test_set_features.csv')
x_test_handcrafted = df_features_labels.loc[test_df['catalogNumber']].drop(columns=['best_tool_encoded'])
y_test_handcrafted = df_features_labels.loc[test_df['catalogNumber'], 'best_tool_encoded']
y_train_handcrafted

best_features = x_train_handcrafted.columns
x_train = x_train_handcrafted
y_train = y_train_handcrafted
x_test = x_test_handcrafted
y_test = y_test_handcrafted
```

```python
# Function for  evaluating trained models
def run_model(model, x_test, y_test):
    preds = model.predict(x_test[list(best_features)])
    errors = preds != y_test
    accuracy = np.mean(preds == y_test)
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Number of errors: {errors.sum()} out of {len(y_test)} ({errors.sum() / len(y_test) * 100:.2f}%)")
    cm = confusion_matrix(y_test, preds)
    cr = classification_report(y_test, preds)
    print("Model Evaluation:")
    print(cm)
    print(cr)
    return accuracy, cm, cr
```

```python
# Hyperparameter optimization with GridSearchCV 
# Uncommented parts show the ranges of hyperparameters evaluated
param_grids = {
    'RandomForest': {
        'n_estimators': [140],
        # 'n_estimators': [int(x) for x in np.linspace(start=55, stop=175, num=8)],
        'max_features': ['sqrt'],
        #'max_features': ['log2', 'sqrt'],
        'max_depth': [None],
        # 'max_depth': [None] + [int(x) for x in np.linspace(10, 100, num=6)],
        'min_samples_split': [10],
        # 'min_samples_split': [5, 10, 20],
        'min_samples_leaf': [1],
        # 'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True],
        # 'bootstrap': [True, False],
        'criterion': ['gini']
        # 'criterion': ['gini', 'entropy']
    },
    'SVC': {
        'C': [100],
        # 'C': [1e-6, 1e-4, 1e-3, 0.01, 0.1, 1, 10, 100],
        'gamma': [0.1],
        # 'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
        'kernel': ['rbf'],
        # 'kernel': ['rbf', 'linear']
        'max_iter': [-1]
        # 'max_iter': [-1, 1, 10, 100, 1000]
    },
    'MLPClassifier': { # Max-Gain: 0.04	
        'hidden_layer_sizes': [42],
        # 'hidden_layer_sizes': [int(x) for x in np.linspace(start=1, stop=50, num=49)],
        'activation': ['relu'],
        # 'activation': ['logistic', 'relu', 'tanh', 'identity'],
        'learning_rate_init': [0.000999],
        # 'learning_rate_init': [0.0001, 0.0005, 0.000999, 0.001, 0.005, 0.01, 0.1],
        'solver': ['adam'],
        # 'solver': ['adam', 'sgd'],
        'batch_size': [16],
        # 'batch_size': [int(x) for x in np.linspace(start=16, stop=128, num=8)],
        # 'early_stopping': [True],
        'alpha': [0.0001875],
        # 'alpha': [float(x) for x in np.linspace(start=0.00010, stop=0.00020, num=9)]
    },
    'KNN': { # Max-Gain: 0.034286
        'n_neighbors': [24],
        # 'n_neighbors': [int(x) for x in np.linspace(start=2, stop=50, num=49)],
        'weights': ['distance'],
        # 'weights': ['uniform', 'distance'],
        'metric': ['manhattan'],
        # 'metric': ['minkowski', 'euclidean', 'manhattan', 'cityblock', 'cosine', 'haversine', 'l1', 'l2', 'nan_euclidean'],
    },
    'LogReg': { # Max-Gain 0.0
        'C': [1.0],
        # 'C': [0.1, 1.0, 10.0],
        'solver': ['liblinear'],
        # 'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'l1_ratio': [0],
        # 'l1_ratio': [0, 1],
    },
    'NB': { # Max-Gain: 0.005714
        # 'var_smoothing': [0.002],
        'var_smoothing': [2e-1, 2e-2, 2e-3, 2e-4, 2e-5, 2e-6, 2e-7]
    },  
    'LightGBM': { # Max-Gain: 0.062857
        'num_leaves': [8],
        # 'num_leaves': [2, 4, 8, 16, 24, 32],
        'n_estimators': [26],
        # 'n_estimators': [int(x) for x in np.linspace(start=2, stop=32, num=31)],
        'max_depth': [-1],
        # 'max_depth': [-1, 0, 2, 4, 6, 8, 10, 15, 20],
        'min_child_samples': [20],
        # 'min_child_samples': [5, 10, 20, 30, 40, 50],
        'learning_rate': [0.09721804511278195], 
        # 'learning_rate': [float(x) for x in np.linspace(start=0.098, stop=0.02, num=400)],
        'reg_alpha': [0.00725],
        # 'reg_alpha': [float(x) for x in np.linspace(start=0.006, stop=0.011, num=5)]
    },
    'GradientBoosting': { # Max-Gain: 0.011429
        'n_estimators': [14],
        # 'n_estimators': [int(x) for x in np.linspace(start=2, stop=100, num=99)],
        'loss': ['log_loss'],
        # 'loss': ['log_loss', 'exponential'],
        'learning_rate': [0.1],
        # 'learning_rate': [float(x) for x in np.linspace(start=0.09, stop=0.11, num=21)],
        'min_samples_split': [20],
        # 'min_samples_split': [2, 5, 10, 20, 30],
        'min_samples_leaf': [19],
        # 'min_samples_leaf': [int(x) for x in np.linspace(start=15, stop=20, num=6)],
        'max_depth': [3],
        # 'max_depth': [None, 3, 5, 10],
        'max_features': ['log2'],
        # 'max_features': ['sqrt', 'log2'], # log2
        'subsample': [0.681] # num=80
        # 'subsample': [float(x) for x in np.linspace(start=0.1, stop=1.0, num=80)] # num=80
    }
}

# Test single model from the param_grids to check if the function works TODO: Remove later
# param_grids = {'RandomForest': param_grids['RandomForest']}
param_grids = {'SVC': param_grids['SVC']}
# param_grids = {'MLPClassifier': param_grids['MLPClassifier']}
# param_grids = {'KNN': param_grids['KNN']}
# param_grids = {'LogReg': param_grids['LogReg']}
# param_grids = {'NB': param_grids['NB']}
# param_grids = {'LightGBM': param_grids['LightGBM']}
# param_grids = {'GradientBoosting': param_grids['GradientBoosting']}

def run_all_models(x_train, y_train, x_test, y_test, param_grids):
    best_models = {}
    callbacks = []
    for model_name, param_grid in param_grids.items():
        print(f"Running GridSearchCV for {model_name}...")
        # Initialize each model
        if model_name == 'RandomForest':
            model = RandomForestClassifier(random_state=42)
        elif model_name == 'SVC':
            model = SVC(random_state=42, probability=True)
        elif model_name == 'MLPClassifier':
            model = MLPClassifier(random_state=42)
        elif model_name == 'CatBoostClassifier':
            model = CatBoostClassifier(random_state=42, verbose=0)
        elif model_name == 'KNN':
            model = KNeighborsClassifier()
        elif model_name == 'LogReg':
            model = LogisticRegression(random_state=42)
        elif model_name == 'NB':
            model = GaussianNB()
        elif model_name == 'LightGBM':
            model = LGBMClassifier(random_state=42)
            callbacks.append(log_evaluation(period=0))
            callbacks.append(early_stopping(stopping_rounds=10))
        elif model_name == 'AdaBoost':
            model = AdaBoostClassifier(random_state=42)
        elif model_name == 'GradientBoosting':
            model = GradientBoostingClassifier(random_state=42)
        grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
        if model_name == 'LightGBM':
            grid_search.fit(x_train[list(best_features)], y_train, callbacks=callbacks, eval_set=[(x_test[list(best_features)], y_test)], eval_metric='binary_logloss')
        else:
            grid_search.fit(x_train[list(best_features)], y_train)
        # print all accuracies for all parameter combinations
        print("All Hyperparameter combinations and their corresponding mean test scores:")
        for mean_score, params in zip(grid_search.cv_results_['mean_test_score'], grid_search.cv_results_['params']):
            print(f"{mean_score:.4f} for {params}")

        print(f"Best Hyperparameters for {model_name}:", grid_search.best_params_)
        best_model = grid_search.best_estimator_
        # best_preds = best_model.predict(x_test[list(best_features)])
        # best_cm = confusion_matrix(y_test, best_preds)
        # best_cr = classification_report(y_test, best_preds)
        # print(best_cm)
        # print(best_cr)
        best_models[model_name] = best_model
        dump(best_model, open(f'../../models/tabular/best_{model_name}_grid.pkl', 'wb')) # save the best model
        results_df = pd.DataFrame(grid_search.cv_results_)
        results_df.to_csv(f'../../models/{model_name}_grid_search_results.csv', index=False) # save cross-validation results
    return best_models
best_models = run_all_models(x_train, y_train, x_test, y_test, param_grids)

# comparison of the trained models with default settings of them
results = []
for model_name, best_model in best_models.items():
    print(f"Evaluating {model_name}...")
    accuracy, cm, cr = run_model(best_model, x_test, y_test)
    default_model = None
    if model_name == 'RandomForest':
        default_model = RandomForestClassifier(random_state=42)
    elif model_name == 'SVC':
        default_model = SVC(random_state=42)
    elif model_name == 'MLPClassifier':
        default_model = MLPClassifier(random_state=42)
    elif model_name == 'CatBoostClassifier':
        default_model = CatBoostClassifier(random_state=42, verbose=0)
    elif model_name == 'KNN':
        default_model = KNeighborsClassifier()
    elif model_name == 'LogReg':
        default_model = LogisticRegression(random_state=42)
    elif model_name == 'NB':
        default_model = GaussianNB()
    elif model_name == 'LightGBM':
        default_model = LGBMClassifier(random_state=42)
    elif model_name == 'AdaBoost':
        default_model = AdaBoostClassifier(random_state=42)
    default_model.fit(x_train[list(best_features)], y_train)
    default_accuracy, default_cm, default_cr = run_model(default_model, x_test, y_test)

    results.append({
        'model': model_name,
        'accuracy': accuracy,
        'accuracy_gain': accuracy - default_accuracy,
        'confusion_matrix': cm,
        'classification_report': cr
    })

results_df = pd.DataFrame(results)
# results_df.to_csv('../../models/model_comparison_results.csv', index=False)
results_df
```
