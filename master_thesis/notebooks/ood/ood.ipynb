```python
#DEFAULT
import sys
sys.path.append("../..")
sys.path.append("../../..")
#3RD PARTY
import pandas as pd
#CUSTOM
import master_thesis.Classes.evaluator as ev
import master_thesis.Classes.aggregator as agg
```

```python
transcriptions_ood = pd.read_csv("../../data/sample_images_ood/transcriptions_ood.csv")
metadata_ood = pd.read_csv("../../data/sample_images_ood/metadata_ood_utf.csv", encoding="UTF-8", sep=";")
```

```python
aggregator = agg.Aggregator(['vouchervision'])
aggregator.add_transcriptions("vouchervision", "../../data/sample_images_ood/transcriptions_ood.csv")
aggregator.aggregate()
res = aggregator.get_aggregated_data()
aggregator.save_aggregated_data("../../data/sample_images_ood/aggregated_data_ood.csv")
res
```

```python
evaluator = ev.Evaluator()
evaluator.load_ground_truth(metadata_ood)
evaluator.add_transcription_results(res)
evaluation_results = evaluator.evaluate_all()
evaluation_results
```

```python
evaluator.export_results(evaluation_results, "../../data/sample_images_ood/evaluation_results_ood.csv")
res_for_export = evaluator.results_for_export
res_for_export
```

```python
records_and_pipelines = []
transcription_results_df = pd.read_csv("../../data/sample_images_ood/evaluation_results_ood.csv")
for index, row in transcription_results_df.iterrows():
    catalog_number = row["catalogNumber"]
    composite_score = row["composite_score"]
    if composite_score >= 0.8:
        recommended_pipeline = "vouchervision"
    else:
        recommended_pipeline = "herbonauten"
    records_and_pipelines.append({
        "catalogNumber": catalog_number, 
        "best_tool": recommended_pipeline})
records_and_pipelines_df = pd.DataFrame(records_and_pipelines)
records_and_pipelines_df.to_csv("../../data/sample_images_ood/labelled_ood_records.csv", index=False)
```
